```python
records = []
for i in range(0, 302):
    records.append([str(data.values[i,j]) for j in [2,11]])
```

```python
records
```

```python
association_rules = apriori(records, min_support=0.3, min_confidence=0.1, min_lift=1, min_length=2)
association_results = list(association_rules)
```
